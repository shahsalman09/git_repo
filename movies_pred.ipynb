```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import the data

df = pd.read_csv('E:/projects/data/tmdb_5000_movies.csv')
df.head()
```

```python
# find columns names

df.columns
```

```python
# drop unnacessary columns

df.drop(columns=['genres','homepage', 'production_companies','production_countries', 'keywords','spoken_languages'], inplace=True, axis=1)
```

```python
df.head(2)
```

```python
# find null values

df.isnull().sum()
```

```python
# impute missing values

df['tagline'].fillna(df['tagline'].mode()[0], inplace=True)
df['overview'].fillna(df['overview'].mode()[0], inplace=True)
df['runtime'].fillna(df['runtime'].mode()[0], inplace=True)
df['release_date'].fillna(df['release_date'].mode()[0], inplace=True)
```
