# 12. Split data into Two subsets

```python
kashti = sns.load_dataset('titanic')
kashti.shape
```

```python
from random import random

kashti_1 = kashti.sample(frac=0.50, random_state=1)
kashti_1.shape
```

```python
kashti_2 = kashti.drop(kashti_1.index)
kashti_2.shape
```

# 13. Join Two datasets

```python
kashti = pd.concat([kashti_1,kashti_2])
kashti
```

# 14. Filtring a dataset

```python
kashti.head()
```

# 19. Reshape multi-index series

```python
import seaborn as sns
df = sns.load_dataset('titanic')
```

```python
df['survived'].mean()
```

```python
df.groupby('sex')['survived'].mean()
```

```python
# Using unstack function

df.groupby(['sex','pclass', 'who'])['survived'].mean().unstack()
```
