# K_Mean Clustering

```python
# imoprt libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# load dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
X = df[['age','fare']]
```

```python
sc = StandardScaler()
scaler =sc.fit_transform(X)
```

```python
model = KMeans(n_clusters=5)
```

```python
model.fit(X)
```

```python
# aasign cluster label to the data
df['cluster'] = model.labels_
```

```python
df['cluster'].unique()
```

```python
plt.scatter(df['age'],df['fare'],c=df['cluster'])
plt.show()
```

```python
import plotly.express as px
fit = px.scatter(df, x='age', y='fare', color='cluster')
fit.show()
```

```python
fit = px.scatter_3d(df, x='age', y='fare', z='pclass',color='cluster')
fit.show()
```

```python
fig = px.parallel_coordinates(df, color='cluster', title='titanic')
fig.show()
```
