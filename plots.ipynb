# Making different plots

```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df['survived'].value_counts().plot(kind="bar")
```

```python
sns.countplot(x='survived', data=df, hue="sex")
```

```python
sns.scatterplot(x='age', y='fare', hue='pclass',data=df)
```

```python
sns.barplot(x='sex',y='fare', hue='pclass',data=df)
```

```python
sns.pairplot(df)
```

```python
df.hist(bins=10, figsize=(9,7), grid=False)
```

```python
sns.boxplot(x='sex', y='fare', hue='who', data=df)
```

```python
sns.boxenplot(x='sex', y='fare', hue='class', data=df)
plt.title('Boxen Plot')
plt.xlabel('Sex')
plt.ylabel('Fare')
```

```python
g = sns.FacetGrid(df, hue='survived', col='pclass', margin_titles=True, palette={1:'seagreen', 0:'gray'})
g = g.map(plt.scatter, 'fare', 'age', edgecolor= 'w').add_legend()
```

```python
sns.stripplot(data=df, x='sex', y='fare', hue='class')
```

```python

numeric_df = df.select_dtypes(include=['float64', 'bool', 'int64'])
corr = numeric_df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
```

```python
sns.violinplot(data=df, x='sex', y='fare', hue='class')
```
